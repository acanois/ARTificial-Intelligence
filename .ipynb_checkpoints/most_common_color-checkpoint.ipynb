```python
import skimage
from skimage import data
from skimage import io
import numpy as np
```

```python
img=io.imread('Rothko_AI/1970_2.jpg')
```

```python
flat_img=[]
for x in img:
    for y in x:
        flat_img.append(y)
```

```python
flat_img=[str(x) for x in flat_img]
counts = {}
for i in flat_img:
    if i in counts:
        counts[i] += 1
    else:
        counts[i] = 1
```

```python
key_max = max(counts.keys(), key=(lambda k: counts[k]))
key_max=key_max.split('[')[1].split(']')[0]
key_max
```

```python
rgb=np.fromstring(key_max, dtype=int, sep=' ')
```

```python
skimage.external.tifffile.imshow(img)
import matplotlib.pyplot as plt
plt.show()
```

```python
skimage.external.tifffile.imshow(np.array([np.array((np.array(rgb),np.array(rgb))),np.array((np.array(rgb),np.array(rgb)))])
)
import matplotlib.pyplot as plt
plt.show()
```
