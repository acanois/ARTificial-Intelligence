```python
import skimage
from skimage import data
from skimage import io
import numpy as np
```

```python
img=io.imread('Rothko_AI/1948_1.jpg')
```

```python
flat_img=[]
for x in img:
    for y in x:
        flat_img.append(y)
```

```python
r = np.array([x[0] for x in flat_img]).mean()
g = np.array([x[1] for x in flat_img]).mean()
b = np.array([x[2] for x in flat_img]).mean()
rgb=[r,g,b]
```
